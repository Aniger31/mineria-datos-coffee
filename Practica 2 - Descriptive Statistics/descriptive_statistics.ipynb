Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargar dataset limpio

```python
df = pd.read_csv("../dataset/coffee_cleaned.csv")
df.head()
```

Estadisticas basicas

```python
df.describe()
```

Promedio de producción

```python
df['production'].mean()
```

Producción por país

```python
production_by_country = df.groupby('country')['production'].mean()
production_by_country.sort_values(ascending=False).head(10)
```

Producción por año

```python
production_by_year = df.groupby('year')['production'].mean()
production_by_year.head()
```

Consumo promedio

```python
df['domestic_consumption'].mean()
```

Relación producción vs consumo

```python
df[['production', 'domestic_consumption']].corr()
```

Gráfica: Top 10 países productores

```python
top10 = production_by_country.sort_values(ascending=False).head(10)

top10.plot(kind='bar')
plt.title("Top 10 países productores de café")
plt.xlabel("País")
plt.ylabel("Producción promedio")
plt.xticks(rotation=45)
plt.show()
```

Gráfica: Producción por año

```python
production_by_year.plot()
plt.title("Producción de café a lo largo del tiempo")
plt.xlabel("Año")
plt.ylabel("Producción promedio")
plt.show()
```

## Conclusiones

- Brasil es el principal productor de café a nivel mundial.
- Existen diferencias significativas entre países productores.
- La producción de café muestra variaciones a lo largo del tiempo.
