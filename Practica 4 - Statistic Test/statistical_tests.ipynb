```python
import pandas as pd
import scipy.stats as stats
```

```python
df = pd.read_csv("../dataset/coffee_cleaned.csv")
df.head()
```

Obtener top 3 países

```python
top_countries = df.groupby('country')['production'].mean().sort_values(ascending=False).head(3).index
top_countries
```

Crear grupos

```python
group1 = df[df['country'] == top_countries[0]]['production']
group2 = df[df['country'] == top_countries[1]]['production']
group3 = df[df['country'] == top_countries[2]]['production']
```

## ANOVA

Se realizó una prueba ANOVA para comparar la producción de café entre los tres países con mayor producción.

El valor p obtenido fue de 1.74e-30, el cual es significativamente menor a 0.05.

Esto indica que existen diferencias estadísticamente significativas entre los grupos analizados, por lo que la producción de café varía entre los países.


```python
anova_result = stats.f_oneway(group1, group2, group3)
anova_result
```

## T-test

Se realizó una prueba T para comparar la producción de café entre dos de los principales países productores.

El p-value obtenido fue de 1.88e-22, el cual es menor a 0.05.

Esto indica que existe una diferencia estadísticamente significativa entre ambos países en términos de producción de café.

```python
ttest_result = stats.ttest_ind(group1, group2)
ttest_result
```

## Conclusiones

Los resultados de las pruebas estadísticas (ANOVA y T-test) demuestran que existen diferencias significativas en la producción de café entre los principales países productores.

Esto confirma que la producción no es homogénea a nivel global y que ciertos países dominan significativamente el mercado.

Estos hallazgos serán útiles para futuras prácticas, especialmente en modelos de predicción y clustering.
