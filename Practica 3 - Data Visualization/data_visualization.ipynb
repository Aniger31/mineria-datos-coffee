```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargar dataset limpio

```python
df = pd.read_csv("../dataset/coffee_cleaned.csv")
df.head()
```

Gráfica 1: Histograma
## Histograma de la producción de café

El histograma muestra la distribución de los valores de producción de café en el dataset.

Se observa la frecuencia con la que aparecen ciertos rangos de producción, lo que permite identificar si los datos están concentrados en valores bajos, medios o altos.

```python
plt.hist(df['production'], bins=20)
plt.title("Distribución de la producción de café")
plt.xlabel("Producción")
plt.ylabel("Frecuencia")
plt.show()
```

Gráfica 2: Boxplot
## Boxplot de la producción de café

El boxplot permite visualizar la distribución de la producción de café, así como identificar valores atípicos (outliers).

Se observan los cuartiles, la mediana y posibles valores extremos que pueden influir en el análisis.

```python
sns.boxplot(x=df['production'])
plt.title("Boxplot de la producción de café")
plt.show()
```

Gráfica 3: Scatter plot
## Relación entre producción y consumo

El gráfico de dispersión muestra la relación entre la producción de café y el consumo doméstico.

Permite identificar si existe alguna correlación entre ambas variables, es decir, si a mayor producción también aumenta el consumo.

```python
plt.scatter(df['production'], df['domestic_consumption'])
plt.title("Producción vs Consumo")
plt.xlabel("Producción")
plt.ylabel("Consumo")
plt.show()
```

Gráfica 4: Línea (tiempo)
## Producción de café a lo largo del tiempo

La gráfica de líneas muestra la evolución de la producción promedio de café a través de los años.

Permite identificar tendencias, incrementos o disminuciones en la producción a lo largo del tiempo.

```python
production_by_year = df.groupby('year')['production'].mean()

production_by_year.plot()
plt.title("Producción de café a lo largo del tiempo")
plt.xlabel("Año")
plt.ylabel("Producción promedio")
plt.show()
```

Gráfica 5: Barras (Top países)
## Top 10 países productores de café

La gráfica de barras muestra los países con mayor producción promedio de café.

Permite comparar fácilmente la producción entre países y visualizar cuáles dominan el mercado.

```python
top10 = df.groupby('country')['production'].mean().sort_values(ascending=False).head(10)

top10.plot(kind='bar')
plt.title("Top 10 países productores")
plt.xlabel("País")
plt.ylabel("Producción")
plt.xticks(rotation=45)
plt.show()
```

Pie chart
## Distribución de la producción (Top 5 países)

La gráfica de pastel representa la proporción de producción de café de los principales países.

Permite observar qué porcentaje del total corresponde a cada país dentro del grupo de mayor producción.

```python
top5 = df.groupby('country')['production'].mean().sort_values(ascending=False).head(5)

top5.plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribución de producción (Top 5 países)")
plt.ylabel("")
plt.show()
```

## Conclusiones

- La producción de café presenta una distribución desigual.
- Existen valores extremos en algunos países.
- Brasil domina la producción global.
- Se observa una tendencia en la producción a lo largo del tiempo.
