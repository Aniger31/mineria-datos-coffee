## Clustering

Se aplicó el algoritmo K-Means para agrupar los datos en diferentes clusters según sus características.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("../dataset/coffee_cleaned.csv")
df.head()
```

```python
X = df[['production', 'domestic_consumption']]
```

## Modelo K-Means

Se utilizó K-Means con 3 clusters para identificar grupos dentro de los datos basados en producción y consumo de café.

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

Obtener clusters

```python
df['cluster'] = kmeans.predict(X)
df.head()
```

## Visualización de clusters

Se generó una gráfica para observar la distribución de los datos y los grupos generados por el modelo.

```python
plt.scatter(df['production'], df['domestic_consumption'], c=df['cluster'])
plt.title("Clustering de producción vs consumo")
plt.xlabel("Producción")
plt.ylabel("Consumo")
plt.show()
```

## Interpretación de los clusters

El modelo K-Means logró identificar tres grupos distintos dentro del dataset.

Se observa un grupo de países con baja producción pero alto consumo, otro grupo con baja producción y bajo consumo, y un tercer grupo con alta producción y consumo medio-alto.

Los clusters están claramente diferenciados en la gráfica, lo que indica que el modelo logró capturar patrones relevantes en los datos.

Centroides (importante)

```python
kmeans.cluster_centers_
```

## Conclusiones

El algoritmo K-Means permitió identificar patrones importantes en la distribución de la producción y el consumo de café.

Los resultados muestran que existen diferentes perfiles de países, desde consumidores hasta grandes productores.

Este tipo de análisis es útil para segmentar datos y comprender mejor el comportamiento del mercado.
