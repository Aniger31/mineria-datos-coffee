## Análisis de texto

Se realizó un análisis de texto utilizando la variable de países para identificar las palabras más frecuentes.

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("../dataset/coffee_cleaned.csv")
df.head()
```

```python
text = " ".join(df['country'].astype(str))
```

## Word Cloud

Se generó una nube de palabras (word cloud) para visualizar la frecuencia de los países dentro del dataset.

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud de países")
plt.show()
```

## Interpretación del Word Cloud

La nube de palabras muestra los países con mayor frecuencia dentro del dataset.

Se observa que algunos países, especialmente de África como Congo y República Centroafricana, aparecen con mayor tamaño, lo que indica una mayor presencia en los datos.

También destacan países como Brasil y Costa Rica, que son relevantes en la producción de café.

Esto sugiere que el dataset tiene una concentración en ciertas regiones geográficas.

## Conclusiones

El análisis de texto permitió identificar los países más representados en el dataset de forma visual e intuitiva.

La word cloud facilita la detección de patrones en la distribución geográfica de los datos.

Este tipo de análisis complementa los métodos numéricos y ayuda a obtener una visión general del dataset.
